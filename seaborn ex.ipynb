```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,)#kde is curve line aka  kernal density estimation
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')#joint plot is for vs plot
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')#plot the every numerical columns in the data
#hue='' for categorical columns
#palette='coolwarm' is for colorise something
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)#barplot is for campare categorical columns with numerical columns
#estimator is a Statistical function to estimate within each categorical bin. by default is mean but we can run our own operation like std
```

```python
sns.countplot(x='sex',data=tips)#count the no.of male and female in data
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
#in this we can is visualize with one numeric column and two categorical columns
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')#it gives us  coefficeint of all the data or kde of all the datapoints
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
# swarm plot strip plot and voilen pilot
# can't scale with large numbers or datasets
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
#this is for explotary data analysis
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

# grids plot

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)# first it gives us blank grids so we can fill it
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
t = sns.FacetGrid(data=tips,col='time',row='smoker')
```

```python
t = sns.FacetGrid(data=tips,col='time',row='smoker')
t.map(sns.distplot,'total_bill')
```

```python
# for more then two argument
t = sns.FacetGrid(data=tips,col='time',row='smoker')
t.map(plt.scatter,'total_bill','tip')
```
